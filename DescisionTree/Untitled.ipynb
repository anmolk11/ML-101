```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv("Purchase_new.csv")
df
```

```python
df.replace(['yes','Yes','no','No'],[1,1,0,0],inplace = True)
```

```python
df
```

```python
X = df.drop(['Purchase'],axis = 'columns')
y = df['Purchase']
```

```python
model = DecisionTreeClassifier()
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
from sklearn import tree
fig, ax = plt.subplots(figsize=(12, 12))
tree.plot_tree(model, ax=ax, feature_names=X.columns, class_names='Purchased')
plt.show()
```
