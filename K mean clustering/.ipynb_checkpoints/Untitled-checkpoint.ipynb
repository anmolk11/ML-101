```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('Mall_Customers.csv')
df
```

```python
sns.scatterplot(data = df,x = 'Annual Income (k$)',y = 'Spending Score (1-100)')
plt.show()
```

```python
model = KMeans(n_clusters=5)
```

```python
clusters = model.fit_predict(df[['Annual Income (k$)','Spending Score (1-100)']])
clusters
```

```python
sns.scatterplot(data = df,x = 'Annual Income (k$)',y = 'Spending Score (1-100)',hue = clusters,palette="deep")
plt.show()
```

```python
centers = model.cluster_centers_
centers
```

```python
sns.scatterplot(data = df,x = 'Annual Income (k$)',y = 'Spending Score (1-100)',hue = clusters,palette="deep")
df_centers = pd.DataFrame(centers, columns=['X', 'Y'])
sns.scatterplot(x='X', y='Y', data=df_centers,marker='X', s=100, color='black')
plt.show()
# X represent clusters center points
```

#### Method for finding a correct value of K (Elbow Tecnique)

```python
# sum of squred error
# sse = model.inertia_
# print(sse)
# the goal is to keep the sse as low as possible
sse_range = []
for k in range(2,10):
    model.fit(df[['Annual Income (k$)','Spending Score (1-100)']])
    sse = model.inertia_
    sse_range.append(sse)
```
